```python
import collections
from itertools import count

def calculate_char_frequency(cipher_text):
    text = cipher_text.lower()
    char_count = collections.Counter(char for char in text if char.isalpha())
    total_chars = sum(char_count.values())

    frequencies = [(char, count, (count/total_chars)*100) for char, count in char_count.items()]

    frequencies.sort(key=lambda x: x[2], reverse=True)
    return frequencies

cipher1_freq = calculate_char_frequency(cipher1)
cipher2_freq = calculate_char_frequency(cipher2)
```

decryption


```python
def decrypt(ciphertext, key):
    plaintext = []
    for char in ciphertext.lower():
        if char.isalpha():
            plaintext.append(key.get(char, '_')) # Use '_' for unmapped characters
        else:
            plaintext.append(char)
    return "".join(plaintext)
```
